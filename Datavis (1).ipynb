```python
df_wb=pd.read_csv('/Users/Utilisateur/Downloads/WDIPCA.csv', sep=",") #Data frame containing the World Bank data
```

```python
df_wb=df_wb.applymap(lambda s:s.lower() if type(s) == str else s)
```

```python
df_wb=df_wb.rename(columns={"Annee":"year", "Short Name":"Country name"})
```

```python
hap_wb= pd.merge(df_hap, df_wb, on=["Country name", "year"])
```

```python
hap_wb.head()
```

```python
#As there are many columns, we are going to select just those that we consider relevant for this work
```

```python
selected_columns=['Country name', 'year', 'Life Ladder', 'Log GDP per capita',
       'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption',
       'F_Adolecent_Fertility_rate‰', 'F_Adolescents_out_of_school%',
       'F_Contributing_family_workers', 'F_Employers',
       'F_Employment_agriculture', 'F_Employment_industry',
       'F_Employment_services',
       'F_Employment in senior and middle management', 'F_Fertility_rate',
       'F_Labor_force', 'F_Life_expectancy',
       'F_Primary_education_teachers', 'Women_national_parliaments',
        'F_Suicide_mortality_rate',
       'Women Business and the Law Index Score',
       'Income Group'] #Pas claire'F_Employment_agriculture' et 'F_Employment_services'
```

```python
selec_df = hap_wb.filter(items = selected_columns)
```

```python
top = selec_df.loc[(selec_df['Life Ladder'] >7) & (selec_df['year']==2019)].sort_values(by='Life Ladder', ascending=False)
```

```python
top.shape
```

```python
top.head(11)
```

```python
flop = selec_df.loc[(selec_df['Life Ladder'] <4) & (selec_df['year']==2019)].sort_values(by='Life Ladder') #2019 because it's the most recent
```

```python
flop.shape
```

```python
ax = top.sort_index(axis=1).plot.bar(x='Country name', y='Life Ladder', width=0.45, color='lightpink', edgecolor = 'hotpink')

ax.set_xlabel("Country", fontsize=16)
ax.set_ylabel("Life Ladder", fontsize=16)
ax.set_title("Top Happiest Countries in 2019", fontsize=18)
ax.tick_params(axis='both', which='major', labelsize=12)
```

```python
ax = flop.sort_index(axis=1).plot.bar(x='Country name', y='Life Ladder', width=0.45, color='lightpink', edgecolor = 'hotpink')

ax.set_xlabel("Country", fontsize=16)
ax.set_ylabel("Life Ladder", fontsize=16)
ax.set_title("Top Unhappiest Countries in 2019", fontsize=18)
ax.tick_params(axis='both', which='major', labelsize=12)
```

```python
correlation = selec_df[['Life Ladder',
       'F_Contributing_family_workers', 'F_Employers',
       'F_Employment_agriculture', 'F_Employment_industry',
       'F_Employment_services',
       'F_Employment in senior and middle management', 'F_Fertility_rate',
       'F_Labor_force', 'F_Life_expectancy',
       'F_Primary_education_teachers', 'Women_national_parliaments',
        'F_Suicide_mortality_rate',
       'Women Business and the Law Index Score']].corr()
plt.figure(figsize=(20, 10))
sns.heatmap(
    correlation,
    fmt=".1%", # Permet d'afficher des pourcentages
    annot=True
)
```
